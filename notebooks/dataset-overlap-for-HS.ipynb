# Data Import

```python
df_march = pd.read_csv('./HiredScore_Mar28_2024.csv')
df_march['Region'] = df_march['country'].map(country_map)
df_march['Datetime'] = pd.to_datetime('2024-03-28')
df_march.head()
```

```python
df_may = pd.read_csv('./Accenture HiredScore Audit - May 2024 v2 (w country).csv')
df_may['Region'] = df_may['Application Country'].map(country_map)
df_may['Datetime'] = pd.to_datetime('2024-05-31')
df_may.head()
```

# Finding overlaps

```python
march_id = df_march['Application ID'].unique()
may_id = df_may['Application ID'].unique()
overlap = list(set(march_id) & set(may_id))
```

```python
print(f'March: {len(march_id):,} application IDs')
print(f'May: {len(may_id):,} application IDs')
print()
print(f'{len(overlap):,} application IDs in common')
```

```python
df_march_only = df_march[~df_march['Application ID'].isin(overlap)]
df_may_only = df_may[~df_may['Application ID'].isin(overlap)]

df_march_overlap = df_march[df_march['Application ID'].isin(overlap)]
df_march_overlap = df_march_overlap.sort_values(by=['Application ID'])
df_march_overlap.reset_index(inplace=True, drop=True)

df_may_overlap = df_may[df_may['Application ID'].isin(overlap)]
df_may_overlap = df_may_overlap.sort_values(by=['Application ID'])
df_may_overlap.reset_index(inplace=True, drop=True)
```

```python
df_march_overlap.rename(columns = {'country':'Country'}, inplace=True)
df_may_overlap.rename(columns = {'Application Country':'Country'}, inplace=True)
```

```python
df_overlap = pd.merge(df_march_overlap, df_may_overlap, on='Application ID', suffixes=('_03', '_05'))
```

# Format and export

```python
c1 = 'background-color: #FDFD96' # yellow
c2 = '' 
```

```python
def font(val):
    return 'font-family: monospace'
```

```python
df1 =  pd.DataFrame(c2, index=df_overlap.index, columns=df_overlap.columns)

for col in feature_cols:
    for col_ in ['03','05','MATCH']:
        df1[f'{col}_{col_}'] = np.where(df_overlap[f'{col}_MATCH'] == 0, c1, c2)
```

```python
df1_processed =  pd.DataFrame(c2, index=df_overlap_processed.index, columns=df_overlap_processed.columns)

for col in processed_cols:
    for col_ in ['03','05','MATCH']:
        df1_processed[f'{col}_{col_}'] = np.where(df_overlap_processed[f'{col}_MATCH'] == 0, c1, c2)
```

```python
differences = df_overlap.style.apply(lambda _: df1, axis=None).applymap(font)
differences_processed = df_overlap_processed.style.apply(lambda _: df1_processed, axis=None).applymap(font)
```

```python
with pd.ExcelWriter("HS_March_May_Overlap.xlsx") as writer:
    differences.to_excel(writer, sheet_name="Original Values", index=False)  
    differences_processed.to_excel(writer, sheet_name="Processed Values", index=False)  
```
