```python
mongo = gmcdb('production', 'utah')
```

```python
proj = ['grade_data.years_of_relevant_experience_insight', 'samurai_json.total_years_of_experience', 
        'samurai_json.parsing_date', 'grade_data.last_final_score_update', '_updated_at']
```

```python
apps = {a['_id']: a for a in mongo.application.find({'req_id': {'$in': list(df['req_id'].unique())}}, proj)}
len(apps)
```

```python
def get_years_of_experience_group(years):
    years_from = 0

    if years is None:
        return 'UNK'
        
    if years == 0 :
        return '0'
        
    for years_to in [0, 1, 2, 5, 8, 12]:
        if years_from < years <= years_to:
            return f'Between {years_from} to {years_to}'
        
        years_from = years_to
            
    return 'greater than 12'
```

```python
df['fixed_years_of_experience'] = df.apply(lambda r: get_years_of_experience_group(apps[r['_id']]['samurai_json']['total_years_of_experience']) if r['_id'] in apps and 'samurai_json' in apps[r['_id']] else None, axis=1).fillna('UNK')
```

```python
df['fixed_years_of_experience'].value_counts()
```

```python
df['app'] = df['_id'].apply(lambda x: apps[x] if x in apps else None)
```

```python
df['fixed_rel_years_of_experience'] = df.apply(lambda r: get_years_of_experience_group(apps[r['_id']]['grade_data']['years_of_relevant_experience_insight']) if r['_id'] in apps and 'grade_data' in apps[r['_id']] else None, axis=1).fillna('UNK')
```
